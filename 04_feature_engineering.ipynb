```python
import pandas as pd

df = pd.read_csv('datasets/credit_train_preprocessed.csv')
df
```

```python
# Debt-to-Income Ratio: Monthly Debt * 12 / Annual Income
df['Debt Income Ratio'] = df['Monthly Debt'] * 12 / df['Annual Income']

# Credit Utilization: Current Credit Balance / Annual Income
df['Credit Utilization'] = df['Current Credit Balance'] / df['Annual Income']

# Delinquency Flag: Convert Months since last delinquent into binary (e.g., delinquent in past year: yes/no)
df['Delinquency Past Year'] = df['Months since last delinquent'] <= 12

df[['Debt Income Ratio', 'Credit Utilization', 'Delinquency Past Year', 'Credit Score Class']]
```

```python
# Export the df2 as a CSV file
df.to_csv("datasets/credit_train_with_features.csv", index=False)
```
