```python
import pandas as pd

df = pd.read_csv("datasets/credit_train.csv")
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.hist(df["Current Loan Amount"], bins=np.arange(0, 1000000, 10000))
plt.show()
```

```python
def remove_outliers(df, column):
    std, mean = df[column].std(), df[column].mean()

    return df[(df[column] < mean + 2 * std) & (df[column] > mean - 2 * std)]

for column in ["Current Loan Amount", "Annual Income", "Monthly Debt", "Years of Credit History", "Months since last delinquent", "Number of Open Accounts", "Number of Credit Problems", "Current Credit Balance", "Maximum Open Credit", "Bankruptcies", "Tax Liens"]:
    df = remove_outliers(df, column)

df
```

```python
df["Credit Score"] = df["Credit Score"].apply(lambda x: x / 10 if x > 1000 else x)
df
```

```python
def clean_years_in_current_job(x):
    if x == "10+ years":
        return 10
    if x == "< 1 year":
        return 0
    return int(x.split()[0])

df["Years in current job"] = df["Years in current job"].apply(clean_years_in_current_job)
df
```
