```python
df = pd.read_csv("datasets/credit_train_clean.csv")
df
```

```python
std, mean = df_coop["Current Loan Amount"].std(), df_coop["Current Loan Amount"].mean()
std, mean
```

```python
# Rescale Function
def rescale(x1, std1, mean1, std2, mean2):
    z = (x1 - mean1) / std1
    return z * std2 + mean2

rescale(548746.0, std, mean, std_coop, mean_coop)
```

```python
for col in ["Current Loan Amount", "Annual Income", "Monthly Debt", "Current Credit Balance", "Maximum Open Credit"]:
    std, mean = df["Current Loan Amount"].std(), df["Current Loan Amount"].mean()
    
    df[col] = df[col].apply(lambda x: rescale(x, std, mean, std_coop, mean_coop))

df
```

```python
# TODO: Please double check if correct
# Generate Column Credit Score Classification based on Credit Score
def credit_score_classification(score):
    if score < 580:
        return 0 # Poor
    elif score < 670:
        return 1 # Fair
    elif score < 740:
        return 2 # Good
    elif score < 800:
        return 3 # Very Good
    else:
        return 4 # Exceptional

df["Credit Score Class"] = df["Credit Score"].apply(credit_score_classification)
df
```

```python
# Drop Unnecessary Columns
df.drop(["Maximum Open Credit", "Credit Score", "Bankruptcies", "Tax Liens", "Number of Credit Problems", "Loan ID", "Customer ID"], axis=1, inplace=True)
df
```
